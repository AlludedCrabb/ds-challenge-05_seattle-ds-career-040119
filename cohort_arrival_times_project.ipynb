```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('data.csv')
```

```python
plt.style.use('bmh')
```

```python
df.columns = ['Day ' + str(x) for x in list(range(1,7))]
```

```python
df
```

```python
#set this to something like the mean
df['Day 1'][6] = '8:45 AM'
```

```python
def fix_time(x):
    return x[0:-3]
```

```python
df = df.applymap(fix_time)
```

```python
def remove_colon(x):
    if len(x) == 4:
        return x[0] + x[2:]
    else:
        return x[:2] + x[3:]
```

```python
df = df.applymap(remove_colon)
```

```python
df = df.astype(int)
```

```python
fig = plt.figure(figsize=(10,8))

for day in df.columns:

    plt.hist(df[day])

plt.title('Cohort Arrival times')
plt.xlabel('Time of morning')
plt.ylabel('People who arrive at that time')
plt.show()
```
